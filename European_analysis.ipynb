#### SCOMPONGO I DATI DEL GRAFICO 1

```python
df = pd.read_csv('match_team_statistics.csv')
teams = df["teamname"].unique()

df = pd.read_csv('match_information.csv')



#Identifico la colonna hometeamname e awayteamname 

data = []

for team in teams:
    df1 = df.query("hometeamname == @team or awayteamname == @team ")
    gol_casa = df.query("hometeamname == @team ")["scorehome"].sum()
    gol_ospite = df.query("awayteamname == @team ")["scoreaway"].sum()
    somma_gol = gol_ospite + gol_casa
    row = [team, gol_casa, gol_ospite, somma_gol]
    data.append(row)



classifica_gol = pd.DataFrame(data, columns = ['Team', 'Goalhome', 'Goalaway','Goaltotal'])

classifica_gol= classifica_gol.sort_values(by=['Goaltotal'], ascending = False)

#Idmax ritorna l'id della riga con la statitistica al massimo (in questo caso Goaltotal), iloc ritorna la riga del dataframe

maxgoals = classifica_gol.iloc[classifica_gol['Goaltotal'].idxmax()]
                    
mingoals = classifica_gol.iloc[classifica_gol['Goaltotal'].idxmin()]
```

#### SCOMPONGO I DATI DEL GRAFICO 2

```python
df = pd.read_csv('match_events.csv')

event_goal_df = pd.DataFrame(df, columns = ['event', 'minute'])

event_goal_df = event_goal_df.query("event == 'Goal' or event == 'OwnGoal' or event== 'GoalOnPenalty'")

event_1 = "'Goal'"
event_2 = "'OwnGoal'"
event_3 = "'GoalOnPenalty'"

event_1_under_40 = len(event_goal_df.query("event==" + event_1 + " and minute < 40"))
event_2_under_40 = len(event_goal_df.query("event==" + event_2 + " and minute < 40"))
event_3_under_40 = len(event_goal_df.query("event==" + event_3 + " and minute < 40"))


event_1_between_40_80 = len(event_goal_df.query("event==" + event_1 + " and minute >= 40 and minute<= 80"))
event_2_between_40_80 = len(event_goal_df.query("event==" + event_2 + " and minute >= 40 and minute<= 80"))
event_3_between_40_80 = len(event_goal_df.query("event==" + event_3 + " and minute >= 40 and minute<= 80"))

event_1_over_80 = len(event_goal_df.query("event==" + event_1 + " and minute > 80"))
event_2_over_80 = len(event_goal_df.query("event==" + event_2 + " and minute > 80"))
event_3_over_80 = len(event_goal_df.query("event==" + event_3 + " and minute > 80"))

#Creo un nuovo dataframe dove vado a inserire i valori appena raggruppati

# initialize list of lists
data = [[event_1,event_1_under_40,event_1_between_40_80,event_1_over_80],
        [event_2,event_2_under_40,event_2_between_40_80,event_2_over_80], 
        [event_3,event_3_under_40,event_3_between_40_80,event_3_over_80]
       ]

event_minute_df = pd.DataFrame(data,columns = ['event','under_40','between_40_80','over_80'])
```

#### SCOMPONGO I DATI DEL GRAFICO 3

```python
df2 = pd.read_csv('match_information.csv')
referee = df2["refereewebname"].unique()

data2 = []


for ref in referee:
    df1 = df2.query("refereewebname == @ref")
    recupero_tot = df2.query("refereewebname == @ref")["injurytime"].sum()
    recupero_final_tournament = df2.query("refereewebname == @ref and roundname == 'final tournament'")["injurytime"].sum()
    recupero_ottavi = df2.query("refereewebname == @ref and roundname == 'eighth finals'")["injurytime"].sum()
    recupero_quarti = df2.query("refereewebname == @ref and roundname == 'quarter finals'")["injurytime"].sum()
    recupero_semifinale = df2.query("refereewebname == @ref and roundname == 'semi finals'")["injurytime"].sum()
    recupero_finale = df2.query("refereewebname == @ref and roundname == 'final'")["injurytime"].sum()
    row = [ref, recupero_tot,recupero_final_tournament,recupero_ottavi,recupero_quarti,recupero_semifinale,recupero_finale]
    data2.append(row)                                  
                                    
    
recupero_arbitri = pd.DataFrame(data2, columns = ['referee', 'recupero_tot', 'recupero_final_tournament','recupero_ottavi','recupero_quarti','recupero_semifinale','recupero_finale'])

recupero_arbitri= recupero_arbitri.sort_values(by=['recupero_tot'], ascending = False)
```

#### SCOMPONGO I DATI DEL GRAFICO 4

```python
df4 = pd.read_csv('match_player_statistics.csv')
Player_name = df4["playersurname"].unique()

df4 = df4.loc[df4["statsname"] != 'Time spent in low activity']
df4 = df4.loc[df4["statsname"] != 'Time spent in medium activity']
df4 = df4.loc[df4["statsname"] != 'Time spent in high activity']
df4 = df4.loc[df4["statsname"] != 'Time spent in very high activity']
df4 = df4.loc[df4["statsname"] != 'Played Time']
df4 = df4.loc[df4["statsname"] != 'Time spent in opposite half']
df4 = df4.loc[df4["statsname"] != 'Position heatmap']
df4 = df4.loc[df4["statsname"] != 'Time spent in penalty area']
df4 = df4.loc[df4["statsname"] != 'Time spent in attacking third']

df4['value'] = df4['value'].astype(float)

data=[]

for name in Player_name:
    df1 = df4.query("playersurname == @name")
    Top_speed = df4.query("playersurname == @name and statsname == 'Top Speed (Km/h)'")["value"].max()
    Number_sprints = df4.query("playersurname == @name and statsname == 'Sprints'")["value"].sum()
    Assists = df4.query("playersurname == @name and statsname == 'Assists'")["value"].sum()
    Goals = df4.query("playersurname == @name and statsname == 'Goals'")["value"].sum()
    Lost_balls = df4.query("playersurname == @name and statsname == 'Lost balls'")["value"].sum()
   
    
    row = [name,Top_speed,Number_sprints,Lost_balls,Goals,Assists]
    data.append(row)

statistiche_df = pd.DataFrame(data, columns = ['Player_name', 'Top_speed','Number_sprints', 'Lost_balls', 'Goals', 'Assists'])

statistiche_df= statistiche_df.sort_values(by=['Goals'],ascending = False).head(10)
```

#### SCOMPONGO I DATI DEL GRAFICO 5

```python
event_cards_df = pd.DataFrame(df, columns = ['event', 'minute'])

event_cards_df = event_cards_df.query("event == 'YellowCard' or event == 'RedCard'")


event_1 = "'YellowCard'"
event_2 = "'RedCard'"


event_1_first_half = len(event_cards_df.query("event==" + event_1 + " and minute <= 45"))
event_2_first_half = len(event_cards_df.query("event==" + event_2 + " and minute <= 45"))

event_1_second_half = len(event_cards_df.query("event==" + event_1 + " and minute > 45 and minute <= 90"))
event_2_second_half = len(event_cards_df.query("event==" + event_2 + " and minute > 45 and minute <= 90"))

event_1_first_extra_time = len(event_cards_df.query("event==" + event_1 + " and minute > 90 and minute <= 105"))
event_2_first_extra_time = len(event_cards_df.query("event==" + event_2 + " and minute > 90 and minute <= 105"))

event_1_second_extra_time = len(event_cards_df.query("event==" + event_1 + " and minute > 105"))
event_2_second_extra_time = len(event_cards_df.query("event==" + event_2 + " and minute > 105"))


data = [[event_1,event_1_first_half,event_1_second_half,event_1_first_extra_time,event_1_second_extra_time],
        [event_2,event_2_first_half,event_2_second_half,event_2_first_extra_time,event_2_second_extra_time], 
       ]

cards_minute_df = pd.DataFrame(data,columns = ['event','first half','second half','first extra time','second extra time'])
```

```python
app= JupyterDash(__name__)

app.layout = html.Div (children =[html.P('Dashboard Euro 2020',style={"font-weight": "bold","text-align": "center", 'fontSize': 35}),
                                  html.Div([
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children ='## Team-Goal'),
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children='Select metrics'),
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children='Checklist'),
                        dcc.Checklist(['Team','Goalhome', 'Goalaway','Goaltotal'],['Team'], id = "checklist"),
                        dcc.Markdown(children='---'),
                        dcc.Graph (id = 'figure_1'),
                        dcc.Markdown(children='---')
                        ]),
     
                            
                                  
                        html.Div([
                        dcc.Markdown(children ='## Minute-Goal'),
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children='Select minute range'),
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children='Radio items'),
                        dcc.RadioItems([{'label': 'Type of goals under 40 minutes', 'value': 'under_40'},
                                        {'label': 'Type of goals between 40 and 80 minutes', 'value': 'between_40_80'},
                                        {'label': 'Type of goals over 80 minutes', 'value': 'over_80'}],'under_40', 
                                       labelStyle={'display': 'block', 'text-align': 'justify'}, id = 'radio'),
                        dcc.Markdown(children='---'),
                        dcc.Graph (id = 'figure_2'),
                        dcc.Markdown(children='---'),
                        ]),
                                  
                                  
                                  
                                  
                        html.Div([
                        dcc.Markdown(children ='## InjuryTime-referee'),
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children='Select round'),
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children='Checklist'),
                        dcc.Checklist([{'label': 'Injury time of final tournament','value':'recupero_final_tournament'},
                                       {'label': 'Injury time round of 16','value':'recupero_ottavi'},
                                       {'label': 'Injury time of quarter-finals','value':'recupero_quarti'},
                                       {'label': 'Injury time of semi-finals','value':'recupero_semifinale'},
                                       {'label': 'Injury time of final','value':'recupero_finale'}],
                                      ['recupero_final_tournament'],labelStyle={'display': 'block', 'text-align': 'justify'},
                                      id = "checklist1"),
    
                        dcc.Markdown(children='---'),
                        dcc.Graph (id = 'figure_3'),
                        dcc.Markdown(children='---'),
                        ]),
                                  
                                  
                                  
                                  
                        html.Div (children =[html.Div([
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children ='## Statistics of the best players in the tournament'),
                        dcc.Markdown(children='---'),
                        dcc.Markdown(children='Choose the metrics in the dropdown menu :'),
                        dcc.Markdown(children='---'),
                        dcc.Dropdown([{'label': 'Top speed','value':'Top_speed'},
                                      {'label': 'Number of sprints','value':'Number_sprints'},
                                      {'label': 'Lost Balls','value':'Lost_balls'},
                                      {'label': 'Assists','value':'Assists'},
                                      {'label': 'Goals','value':'Goals'}],
                                      ['Goals'], multi = True,
                                      id = "Dropdown"),

                        dcc.Graph (id = 'figure_4'),
                        dcc.Markdown(children='---'),
                        ]),
                                             
                                             
                         
                                             
                         html.Div (children =[html.Div([
                         dcc.Markdown(children='---'),
                         dcc.Markdown(children ='# Cards-Minute'),
                         dcc.Markdown(children='---'),
                         dcc.Markdown(children='Select the time'),
                         dcc.Markdown(children='---'),
                         dcc.Markdown(children='Checklist'),
                         dcc.Checklist([{'label': 'First Half','value':'first half'},
                                        {'label': 'Second half','value':'second half'},
                                        {'label': 'First extra time','value':'first extra time'},
                                        {'label': 'Second extra time','value':'second extra time'},
                                        ],
                            ['first half'],labelStyle={'display': 'block', 'text-align': 'justify'},
                            id = "checklist5"),

                         dcc.Markdown(children='---'),
                         dcc.Graph (id = 'figure_5'),
                         dcc.Markdown(children='---'),
                         ]),

                                             
                                             
                                             
                                             
                                             
                        ])
                        ])
                        ])

    





@app.callback( 
Output(component_id='figure_1', component_property='figure'),
Input(component_id='checklist', component_property='value'),
)

def update_figure_1(checklist):
    fig = px.sunburst(classifica_gol,
                  path = checklist , 
                  values = 'Goaltotal' 
                )
                    
    fig.update_layout(transition = {'duration' : 1000}) 
    return fig




                                  
                                  

@app.callback( 
Output(component_id='figure_2', component_property='figure'),
Input(component_id='radio', component_property='value'),
)

def update_figure_2(radio):
    #style layout 
    layout = go.Layout(
    xaxis=dict(title="Number of goal for range"),
    yaxis=dict(title="Events") 
    ) 

    fig = go.Figure(layout = layout,data=go.Scatter(
        x= event_minute_df[radio],
        y=event_minute_df["event"],
        mode='markers',
        marker=dict(size=[40, 60, 80, 100], color=[0, 1, 2, 3])
        ))
                                  
    fig.update_layout(transition = {'duration' : 1000}) 
    return fig


                                  
                                  
                                  

@app.callback(
Output(component_id='figure_3', component_property='figure'),
Input(component_id='checklist1', component_property='value'),
)

def update_figure_3(checklist1):
    fig = px.bar(recupero_arbitri,
    x = "referee",
    y = checklist1,
              
    labels={"referee" : "Referee",
            "value" : "Minutes"},
                )
    
    fig.update_layout(transition = {'duration' : 1000})
    return fig
                                  
                                  
                
                                  
                                  
                                  
@app.callback(
Output(component_id='figure_4', component_property='figure'),
Input(component_id='Dropdown', component_property='value'),
)

def update_figure_4(Dropdown):
    data =[]
    for column in Dropdown:
        bar = go.Bar(name= column, x=statistiche_df["Player_name"], y=statistiche_df[column])
        data.append(bar)  

    fig = go.Figure(data)
    fig.update_layout(transition = {'duration' : 1000})
    return fig 

        
                                  
        
                                  
                                  
@app.callback(
Output(component_id='figure_5', component_property='figure'),
Input(component_id='checklist5', component_property='value'),
)

def update_figure_5(checklist5):
                                  
    fig = px.funnel(cards_minute_df, y='event', x=checklist5)
    fig.update_layout(transition = {'duration' : 1000}) 
    return fig

                                  
                                  
                                  
                                  

app.run_server(mode = 'inline')
```
